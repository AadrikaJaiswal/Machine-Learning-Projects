```python
#Loading the dataset to a Pandas dataFrame
data= pd.read_csv(r'C:\Users\KIIT\projects\Train price prediction\data1.csv')
df= data.copy()
```

```python
from sklearn.preprocessing import OrdinalEncoder
org=OrdinalEncoder()
dest=OrdinalEncoder()
tty=OrdinalEncoder()
tcl=OrdinalEncoder()
df['origin']=org.fit_transform(df[['origin']])
df['destination']=dest.fit_transform(df[['destination']])
df['train_type']=tty.fit_transform(df[['train_type']])
df['train_class']=tcl.fit_transform(df[['train_class']])
```
